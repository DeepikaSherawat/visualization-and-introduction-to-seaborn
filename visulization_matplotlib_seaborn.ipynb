# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mtcars=pd.read_csv("mtcars.csv")
```

```python
mtcars.head()
```

```python
mtcars.gear.value_counts()
```

```python
mtcars.cyl.value_counts()   #value_count can be used for singe categorical variable
#and crosstab used for multiple categorical variable
```

```python
#table          
pd.crosstab(mtcars.gear,mtcars.cyl)
```

```python
#Bar plot for single category
plt.figure(figsize=(10,4))
mtcars["gear"].value_counts().plot(kind="pie",autopct="%1.1f%%")
```

```python
plt.figure(figsize=(10,4))
mtcars["gear"].value_counts().plot(kind="barh")
```

```python
plt.figure(figsize=(10,4))
mtcars["gear"].value_counts().plot(kind="bar")
```

```python
pd.crosstab(mtcars.gear,mtcars.cyl).plot(kind="bar")
```

```python
plt.scatter(mtcars.mpg,mtcars.wt) #scatter plot of two variables
plt.xlabel("wt")
plt.ylabel("mpg")
plt.title("correlation plot")
```

```python
plt.figure(figsize=(6,6))
plt.hist(mtcars["mpg"],facecolor="black",edgecolor="red",bins=5)
#creates histogram with 5bins and colours filled init.
plt.xlabel("mpg")
plt.title("histogram plot")
```

```python
plt.figure(figsize=(6,6),facecolor="white")
plt.hist(mtcars["hp"],facecolor="black",edgecolor="red",bins=8)
#creates histogram with 5bins and colours filled init.
plt.xlabel("hp")

#outlier preest in data and histogram help us to detect outlier
```

```python
#boxplot
#help(plt.boxplot)
plt.figure(figsize=(20,8),facecolor="lightgreen")
plt.boxplot(mtcars["hp"],vert=False)
plt.ylabel("mpg",size=20)
plt.xlabel("Boxplot",size=20)
plt.title("Boxplot",size=20)  #for vertical
```

```python
#violin plot
#help(plt.violin plot)
plt.figure(figsize=(8,6),facecolor="lightgreen")
plt.violinplot(mtcars["hp"],vert=False)

#it is alternative of histogram plot but not of industry standered thats why never used.
#if we see violin plot for hp there is no data b/w 250 and 300 but violin plot  show data so it can mislead.
```
